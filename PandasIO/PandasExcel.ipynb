# FAZENDO A LEITURA DE ARQUIVOS XLSX

```python
import pandas as pd
```

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True'
```

```python
dados = pd.read_excel(url)
dados
```

```python
# Descobrindo o nome das panilhas presentes no arquivo excel
arquivo_excel = pd.ExcelFile(url)
arquivo_excel.sheet_names
```

Parametros da função pd.read_excel
- sheet_name= --> Qual planilha utilizar
- usecols='...' --> Quais colunas da planilha utilizar (Intervalo)
- nrows= --> Mostra somente o número de linhas selecionadas


sheet_name=

```python
percapita = pd.read_excel(url, sheet_name='emissoes_percapita')
percapita
```

usecols=' '

```python
intervalo = pd.read_excel(url,sheet_name='emissoes_C02', usecols='A:D')
intervalo
```

nrows=

```python
intervalo_2 = pd.read_excel(url, sheet_name='emissoes_C02', usecols='A:D', nrows=10)
intervalo_2
```

Escrevendo arquivos em Excel

```python
percapita.to_excel('percapita.xlsx', index=False)
```
