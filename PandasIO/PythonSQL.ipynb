```python
# Biblioteca para gerar um banco de dados local
import sqlalchemy
```

```python
# Importando coisas uteis
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
# Utilizando banco de dados local
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
dados = pd.read_csv(url)
dados
```

```python
# Funcao para enviar os dados para dentro do SQL
dados.to_sql('clientes', engine, index=False)
```

```python
# Fazendo a inspecao dos dados dentro do banco de dados
inspector = inspect(engine)
```

```python
# Printando as tabelas
print(inspector.get_table_names())
```

Realizando consultas em SQL

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
# Lendo as consultas read_sql(query, banco)
empregados = pd.read_sql(query, engine)
empregados
```

```python
# Salvnado as consultas
empregados.to_sql('empregados',con=engine,index=False)
```

```python
# Codigo para ler somente uma tabela
pd.read_sql_table('empregados',engine)
```

```python
# Lendo uma tabela especificando colunas
pd.read_sql_table('empregados',engine, columns=['ID_Cliente','Grau_escolaridade','Rendimento_anual'])
```

Atualizando dados dentro do banco de dados

```python
# Funcao para garantir que a string seja interpretada como SQL puro
from sqlalchemy import text
```

```python
# Excluindo registros
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
with engine.connect() as conn:
    result = conn.execute(text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes',engine)
```

```python
# Atualizando registros
query = 'UPDATE clientes SET Grau_escolaridade = "Ensino Superior" WHERE ID_Cliente = 5008808	'
with engine.connect() as conn:
    result = conn.execute(text(query))
    conn.commit()
```

```python
# Lendo os registros
pd.read_sql_table('clientes',engine)
```
