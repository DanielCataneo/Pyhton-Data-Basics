```python
import pandas as pd
```

```python
populacao_estados = pd.read_excel('/content/drive/MyDrive/POP2022_Municipios.xls', header=1, skipfooter=34)
populacao_estados
```

```python
populacao_estados.groupby('UF').sum(numeric_only=True)
```

```python
# Transformando a coluna população em numeros
populacao_estados['POPULAÇÃO'].astype(int)
# Erro por conta de números com parenteses presentes na tabela
```

```python
populacao_estados[populacao_estados['POPULAÇÃO'].str.contains('\(', na = False)]
```

TRATANDO OS DADOS para retirar os parenteses e os separadores de milhar

Metodo ASIGN('Nome Coluna' , condicao)

Metodo REPLACE(['Nome da Coluna'].replace('o que procura', 'o que sera substituido')

```python
populacao_estados = populacao_estados.assign(populacao_sem_parenteses = populacao_estados['POPULAÇÃO'].replace('\(\d{1,2}\)', '', regex = True),
                                             populacao = lambda x : x.loc[:, 'populacao_sem_parenteses'].replace('\.', '', regex = True))
```

```python
populacao_estados[populacao_estados['POPULAÇÃO'].str.contains('\(', na = False)]
```

```python
# Transformando populaçao em Int
populacao_estados = populacao_estados.astype({'populacao':'int64'})
```

```python
# Utilizando reset_index para que UF não se torne o index da tabela
populacao_estados = populacao_estados.groupby('UF')['populacao'].sum().reset_index()
populacao_estados
```

```python
populacao_estados.to_csv('Populacao_Totais_Por_Estado.csv',index=False,sep=';')
```
