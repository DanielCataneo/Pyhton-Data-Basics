# LENDO ARQUIVOS CSV COM PANDAS

```python
# CSV comma separated values
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/superstore_data.csv'
```

```python
# importando dados com csv
dados_mercado = pd.read_csv(url)
dados_mercado.head()
```

```python
url_2 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/superstore_data_ponto_virgula.csv'
```

```python
# caso arquivo separado por ;
dados_mercado_virgula = pd.read_csv(url_2, sep=';')
dados_mercado_virgula.head()
```

Parametros da função pd.read_csv()
- nrows= --> Mostra a quantidade de linhas determinada pelo usuário
- sep= --> delimiter dos dados dentro do arquivo (padrão ',')
- usecols[ ] --> Mostra somente as colunas determinadas pelo usuário
obs.: Pode colocar o nome ou somente o numero da posição

Nrows=

```python
dados_mercado_10 = pd.read_csv(url, nrows=10)
dados_mercado_10
```

UseCols[ ]

```python
dados_colunas = pd.read_csv(url, usecols=['Id', 'Year_Birth', 'Income'])
dados_colunas.head()
```

```python
# Sintaxe alternativa com posição das colunas
dados_colunas_2 = pd.read_csv(url, usecols=[0, 1, 4])
dados_colunas_2.head()
```

Salvando arquivo em csv

```python
# Index False para manter somente os indexes padrão das colunas
dados_colunas.to_csv('clientes_mercado.csv', index=False)
```
